### Analysis


Observations:

- Mice treated with Capomulin and Ramicane tend to have significantly lower volume tumors (max 41.5 mm3) compared to mice that were treated with other drugs (50 mm3 or more).
- Out of four drug regimens checked, Infubinol is the only drug to have potential outliers.
- At 0.84, there is a strong correlation between mouse weight and the average tumor volume. 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_combined = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
mouse_combined
```

```python
# Checking the number of mice.
mouse_unique = mouse_combined["Mouse ID"].nunique()
mouse_unique
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse = mouse_combined.loc[mouse_combined.duplicated(subset=["Mouse ID","Timepoint"]), "Mouse ID"].unique()
duplicate_mouse

```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouse_combined[mouse_combined["Mouse ID"] == duplicate_mouse[0]]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# mouse_combined = mouse_combined.drop_duplicates(subset=["Mouse ID","Timepoint"])
# mouse_combined

# Find duplicate then drop all info
clean_mouse_df = mouse_combined.loc[mouse_combined["Mouse ID"].isin(duplicate_mouse)==False]
clean_mouse_df
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_unique = clean_mouse_df["Mouse ID"].nunique()
mouse_unique
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

## Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
summary_group = clean_mouse_df.groupby('Drug Regimen')


## mean, median, variance, standard deviation, and SEM of the tumor volume. 

mean_tumor = summary_group["Tumor Volume (mm3)"].mean()
median_tumor = summary_group["Tumor Volume (mm3)"].median()
variance_tumor = summary_group["Tumor Volume (mm3)"].var()
stddev_tumor = summary_group["Tumor Volume (mm3)"].std()
stderr_tumor = summary_group["Tumor Volume (mm3)"].sem()

## Assemble the resulting series into a single summary dataframe.

drug_summary_df = pd.DataFrame({
    "Mean Tumor Volume": mean_tumor,
    "Median Tumor Volume": median_tumor,
    "Tumor Volume Variance": variance_tumor,
    "Tumor Volume Std. Dev": stddev_tumor,
    "Tumor Volume Std. Err.": stderr_tumor
})

drug_summary_df
```

```python
# Using the aggregation method, produce the same summary statistics in a single line

summary_group["Tumor Volume (mm3)"].aggregate(['mean','median','var','std','sem'])
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
four_drugs = clean_mouse_df.loc[clean_mouse_df['Drug Regimen'].isin(drug_list)]

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = clean_mouse_df.groupby(['Mouse ID'])["Timepoint"]
last_timepoint = last_timepoint.max()
last_timepoint = last_timepoint.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
last_timepoint = last_timepoint.merge(clean_mouse_df, on=["Mouse ID", "Timepoint"], how="left")
last_timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]


# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# For loop to append drug regimen and tumor key value pair in empty list
for drug in drug_list:
    vol = last_timepoint.loc[last_timepoint["Drug Regimen"] == drug]["Tumor Volume (mm3)"]
    tumor_vol.append(vol)

    # Calculate the IQR and quantitatively determine if there are any potential outliers. 
    quartiles = vol.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    iqr_1_5 = iqr * 1.5
    lower_bound = lowerq - iqr_1_5
    upper_bound = upperq + iqr_1_5
    
    # Determine outliers using upper and lower bounds
    outliers = (vol.loc[(last_timepoint['Tumor Volume (mm3)'] > upper_bound) | 
                                        (last_timepoint['Tumor Volume (mm3)'] < lower_bound)])
    print(f"{drug}'s potential outliers: {outliers}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
plt.boxplot(tumor_vol, labels=drug_list, flierprops=dict(markerfacecolor="red"))
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

# Dataframe for Mouse S710
capomulin = ["Capomulin"]
s710list = ["s710"]
cap_df = clean_mouse_df.loc[clean_mouse_df['Drug Regimen'].isin(capomulin)]
s710 = cap_df.loc[cap_df['Mouse ID'].isin(s710list)]


# Line plot for Mouse S710 treatment
cap_tp = s710["Timepoint"]
cap_vol = s710["Tumor Volume (mm3)"]
plt.plot(cap_tp, cap_vol)
plt.title('Capomulin treatment of mouse S710')
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
cap_mean = cap_df.groupby('Mouse ID').mean()

x_tumor_weight = cap_mean["Weight (g)"]
y_tumor_weight = cap_mean["Tumor Volume (mm3)"]

plt.scatter(x_tumor_weight, y_tumor_weight)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

correlation_coefficient = st.pearsonr(x_tumor_weight, y_tumor_weight)
correlation_coefficient = round(correlation_coefficient[0],2)

# Linear regression equation
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_tumor_weight, y_tumor_weight)
regress_values = x_tumor_weight * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot the linear regression model on top of the previous scatter plot
print(f'The correlation between mouse weight and the average tumor volume is {correlation_coefficient}')
plt.scatter(x_tumor_weight,y_tumor_weight)
plt.plot(x_tumor_weight,regress_values,"r-")

plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
